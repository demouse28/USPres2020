```python
# import the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# scrape for the polling data
polls_res = requests.get('https://en.wikipedia.org/wiki/Statewide_opinion_polling_for_the_2008_United_States_presidential_election')
polls_soup = BeautifulSoup(polls_res.content, 'lxml')
```

```python
# create a list of dataframes
dfs = [pd.DataFrame() for i in range(len(polls_soup.find_all('table', {'class': 'wikitable collapsible collapsed'})))]
for i in range(len(polls_soup.find_all('table', {'class': 'wikitable collapsible collapsed'}))):
    dfs[i] = pd.read_html(str(polls_soup.find_all('table', {'class': 'wikitable collapsible collapsed'})[i]))[0]
```

```python
# assign each dataframe(s) to each state
AL = pd.concat([dfs[0], dfs[1]], ignore_index = True)
AK = pd.concat([dfs[2], dfs[3]], ignore_index = True)
AZ = pd.concat([dfs[4], dfs[5]], ignore_index = True)
AR = pd.concat([dfs[6], dfs[7]], ignore_index = True)
CA = pd.concat([dfs[8], dfs[9]], ignore_index = True)
CO = pd.concat([dfs[10], dfs[11], dfs[12]], ignore_index = True)
CT = pd.concat([dfs[13], dfs[14]], ignore_index = True)
DE = dfs[15]
DC = dfs[16]
FL = pd.concat([dfs[17], dfs[18], dfs[19], dfs[20]], ignore_index = True)
GA = pd.concat([dfs[21], dfs[22], dfs[23]], ignore_index = True)
HI = dfs[24]
ID = pd.concat([dfs[25], dfs[26]], ignore_index = True)
IL = pd.concat([dfs[27], dfs[28]], ignore_index = True)
IN = pd.concat([dfs[29], dfs[30], dfs[31]], ignore_index = True)
IA = pd.concat([dfs[32], dfs[33]], ignore_index = True)
KS = dfs[34]
KY = pd.concat([dfs[35], dfs[36]], ignore_index = True)
LA = pd.concat([dfs[37], dfs[38]], ignore_index = True)
ME = pd.concat([dfs[39], dfs[40], dfs[41]], ignore_index = True)
MD = pd.concat([dfs[42], dfs[43]], ignore_index = True)
MA = pd.concat([dfs[44], dfs[45], dfs[46], dfs[47]], ignore_index = True)
MI = pd.concat([dfs[48], dfs[49], dfs[50], dfs[51]], ignore_index = True)
MN = pd.concat([dfs[52], dfs[53]], ignore_index = True)
MS = dfs[54]
MO = pd.concat([dfs[55], dfs[56], dfs[57], dfs[58]], ignore_index = True)
MT = pd.concat([dfs[59], dfs[60], dfs[61]], ignore_index = True)
NE = pd.concat([dfs[62], dfs[63]], ignore_index = True)
NV = pd.concat([dfs[64], dfs[65], dfs[66], dfs[67]], ignore_index = True)
NH = pd.concat([dfs[68], dfs[69], dfs[70]], ignore_index = True)
NJ = pd.concat([dfs[71], dfs[72]], ignore_index = True)
NM = pd.concat([dfs[73], dfs[74], dfs[75], dfs[76]], ignore_index = True)
NY = pd.concat([dfs[77], dfs[78]], ignore_index = True)
NC = pd.concat([dfs[79], dfs[80], dfs[81]], ignore_index = True)
ND = dfs[82]
OH = pd.concat([dfs[83], dfs[84], dfs[85], dfs[86], dfs[87]], ignore_index = True)
OK = pd.concat([dfs[88], dfs[89]], ignore_index = True)
OR = pd.concat([dfs[90], dfs[91], dfs[92]], ignore_index = True)
PA = pd.concat([dfs[93], dfs[94]], ignore_index = True)
RI = dfs[95]
SC = pd.concat([dfs[96], dfs[97], dfs[98]], ignore_index = True)
SD = dfs[99]
TN = pd.concat([dfs[100], dfs[101]], ignore_index = True)
TX = pd.concat([dfs[102][dfs[102]['Barack Obama'] != 'Not reported'], dfs[103], dfs[104]], ignore_index = True)
UT = pd.concat([dfs[105], dfs[106]], ignore_index = True)
VT = dfs[107]
VA = pd.concat([dfs[108], dfs[109], dfs[110], dfs[111]], ignore_index = True)
WA = pd.concat([dfs[112], dfs[113]], ignore_index = True)
WV = pd.concat([dfs[114], dfs[115]], ignore_index = True)
WI = pd.concat([dfs[116], dfs[117]], ignore_index = True)
WY = dfs[118]
```

```python
# aggregate the dataframes into a list of dataframes in state order
dfs = [AL, AK, AZ, AR, CA, CO, CT, DE, DC, FL, GA, HI, ID, IL, IN, IA, KS, KY, LA, ME, MD, MA, MI, MN, MS, MO, MT, NE, NV, NH, NJ, NM, NY, NC, ND, OH, OK, OR, PA, RI, SC, SD, TN, TX, UT, VT, VA, WA, WV, WI, WY]
```

```python
# find the state abbreviations
state_res = requests.get('https://www.ssa.gov/international/coc-docs/states.html')
state_soup = BeautifulSoup(state_res.content, 'lxml')
state_html = state_soup.find(text = 'Two-Letter State Abbreviations').find_next('table')
state = pd.DataFrame(pd.read_html(str(state_html))[0])
state.drop(state[state[0].isin(['AMERICAN SAMOA', 'GUAM', 'NORTHERN MARIANA IS', 'PUERTO RICO', 'VIRGIN ISLANDS'])].index, inplace = True)
state = state.reset_index(drop = True)[1]
state
```

```python
# filter the dataframe to only include the state, month, year, democratic percentage, republican percentage, party winner, margin
polls_all = pd.DataFrame()
for i in range(len(dfs)):
    dfs[i]['state'] = [state[i]] * len(dfs[i])
    month = []
    year = []
    for j in dfs[i]['Date administered (2008)']:
        if '–' in j:
            if any(j.split('–')[1][0] in x for x in [str(x) for x in list(range(10))]):
                month.append(j.split('–')[0][0:3])
            else:
                month.append(j.split('–')[1][0:3])
        if '–' not in j:
            month.append(j.split(' ')[0][0:3])
        if '2007' in j:
            year.append(int(j.split(' ')[-1].split('[')[0]))
        if '2007' not in j:
            year.append(int(2008))
    dfs[i]['month'] = month
    dfs[i]['year'] = year
    dem_percentage = []
    for j in dfs[i]['Barack Obama']:
        dem_percentage.append(float(str(j).split('%')[0]))
    dfs[i]['dem_percentage'] = dem_percentage
    gop_percentage = []
    for j in dfs[i]['John McCain']:
        gop_percentage.append(float(str(j).split('%')[0]))
    dfs[i]['gop_percentage'] = gop_percentage
    winner = []
    margin = []
    for j in range(len(dfs[i])):
        if dfs[i]['dem_percentage'][j] > dfs[i]['gop_percentage'][j]:
            winner.append('D')
        if dfs[i]['dem_percentage'][j] < dfs[i]['gop_percentage'][j]:
            winner.append('R')
        if dfs[i]['dem_percentage'][j] == dfs[i]['gop_percentage'][j]:
            winner.append('E')
        margin.append(abs(dfs[i]['dem_percentage'][j] - dfs[i]['gop_percentage'][j]))
    dfs[i]['winner'] = winner
    dfs[i]['margin'] = margin
    polls_all = pd.concat([polls_all, dfs[i]], ignore_index = True)
```

```python
polls_dr = polls_all[['state', 'month', 'year', 'dem_percentage', 'gop_percentage', 'winner', 'margin']]
polls_dr
```

```python
# filter the dataframe to only include nov 2019 to nov 2020
polls_oneyear = polls_dr[(polls_dr['year'] == 2008) | ((polls_dr['year'] == 2007) & polls_dr['month'].isin(['Nov', 'Dec']))].reset_index(drop = True)
polls_oneyear
```

```python
# negative represents democratic, positive represents republican
party_margin = []
for i in range(len(polls_oneyear)):
    if polls_oneyear['winner'][i] == 'D':
        party_margin.append(-polls_oneyear['margin'][i])
    if any(polls_oneyear['winner'][i] in x for x in ['E', 'R']):
        party_margin.append(polls_oneyear['margin'][i])
polls_party_margin = polls_oneyear
polls_party_margin['2008_polls_margin'] = party_margin
polls_party_margin
```

```python
# find the unweighted mean
polls_mean = polls_party_margin.groupby('state', sort = False)[['2008_polls_margin']].mean().reset_index(drop = False)
polls_mean
```

```python
# export the dataframe to a csv file
polls_mean.to_csv('../../data/processed/2008_mean_polls.csv', index = False)
```
