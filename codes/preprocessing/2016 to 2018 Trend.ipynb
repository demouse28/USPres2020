```python
# import libraries
import pandas as pd
```

```python
# import the 2016 presidential election data
pres_2016 = pd.read_csv('../data/processed/2016_presidential_election.csv')
pres_2016
```

```python
# import the 2018 house election data
house_2018 = pd.read_csv('../data/processed/2018_house_election.csv')
house_2018
```

```python
# create the dataframe for the trend
trend_16_18 = pd.DataFrame(pres_2016['state'])
trend_16_18['2016_winner'] = pres_2016['winner']
trend_16_18['2016_margin_percentage'] = pres_2016['margin_percentage']
trend_16_18['2018_winner'] = house_2018['winner']
trend_16_18['2018_margin_percentage'] = house_2018['margin_percentage']
trend_16_18
```

```python
# create a column to indicate whether the state flipped in the 2018 house election with respect to the 2016 presidential election
flip = []
for i in range(len(trend_16_18)):
    if trend_16_18['2016_winner'][i] == trend_16_18['2018_winner'][i]:
        flip.append(0)
    else:
        flip.append(1)
trend_16_18['flip'] = flip
trend_16_18
```

```python
# create a column to indicate which party does each state trend from the 2016 house election to the 2018 presidential election
trend_party = []
for i in range(len(trend_16_18)):
    if trend_16_18['flip'][i] == 0:
        if trend_16_18['2018_margin_percentage'][i] > trend_16_18['2016_margin_percentage'][i]:
            trend_party.append(trend_16_18['2018_winner'][i])
        if trend_16_18['2018_margin_percentage'][i] < trend_16_18['2016_margin_percentage'][i]:
            if trend_16_18['2018_winner'][i] == 'R':
                trend_party.append('D')
            if trend_16_18['2018_winner'][i] == 'D':
                trend_party.append('R')
    if trend_16_18['flip'][i] == 1:
        trend_party.append(trend_16_18['2018_winner'][i])
trend_16_18['trend_party'] = trend_party
trend_16_18
```

```python
# create a column of the trend percentages for each state from the 2016 house election to the 2018 presidential election
trend_percentage = []
for i in range(len(trend_16_18)):
    if trend_16_18['flip'][i] == 0:
        trend_percentage.append(abs(trend_16_18['2018_margin_percentage'][i] - trend_16_18['2016_margin_percentage'][i]))
    if trend_16_18['flip'][i] == 1:
        trend_percentage.append(trend_16_18['2018_margin_percentage'][i] + trend_16_18['2016_margin_percentage'][i])
trend_16_18['trend_percentage'] = trend_percentage
trend_16_18
```

```python
# export the dataframe to a csv file
trend_16_18.to_csv('../data/processed/2016_to_2018_trend.csv', index = False)
```
