```python
# import the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# scrape for the polling data
# sep 2012 - nov 2012
polls_res1 = requests.get('https://en.wikipedia.org/wiki/Statewide_opinion_polling_for_the_2012_United_States_presidential_election')
polls_soup1 = BeautifulSoup(polls_res1.content, 'lxml')
# jan 2012 - aug 2012
polls_res2 = requests.get('https://en.wikipedia.org/wiki/Early/Mid_2012_statewide_opinion_polling_for_the_2012_United_States_presidential_election')
polls_soup2 = BeautifulSoup(polls_res2.content, 'lxml')
# pre 2012
polls_res3 = requests.get('https://en.wikipedia.org/wiki/Pre-2012_statewide_opinion_polling_for_the_2012_United_States_presidential_election')
polls_soup3 = BeautifulSoup(polls_res3.content, 'lxml')
```

```python
# sep 2012 - nov 2012
dfs1 = [pd.DataFrame() for i in range(len(polls_soup1.find_all('table', {'class': 'wikitable'})))]
for i in range(len(polls_soup1.find_all('table', {'class': 'wikitable'}))):
    dfs1[i] = pd.read_html(str(polls_soup1.find_all('table', {'class': 'wikitable'})[i]))[0]
# jan 2012 - aug 2012
dfs2 = [pd.DataFrame() for i in range(len(polls_soup2.find_all('table', {'class': 'wikitable'})))]
for i in range(len(polls_soup2.find_all('table', {'class': 'wikitable'}))):
    dfs2[i] = pd.read_html(str(polls_soup2.find_all('table', {'class': 'wikitable'})[i]))[0]
# pre 2012
dfs3 = [pd.DataFrame() for i in range(len(polls_soup3.find_all('table', {'class': 'wikitable'})))]
for i in range(len(polls_soup3.find_all('table', {'class': 'wikitable'}))):
    dfs3[i] = pd.read_html(str(polls_soup3.find_all('table', {'class': 'wikitable'})[i]))[0]
```

```python
# aggregate the dataframes into a list of dataframes in state order
# remove states without polls, i.e. AK, DE, WY
dfs = [AL, AZ, AR, CA, CO, CT, DC, FL, GA, HI, ID, IL, IN, IA, KS, KY, LA, ME, MD, MA, MI, MN, MS, MO, MT, NE, NV, NH, NJ, NM, NY, NC, ND, OH, OK, OR, PA, RI, SC, SD, TN, TX, UT, VT, VA, WA, WV, WI]
```

```python
# find the state abbreviations
state_res = requests.get('https://www.ssa.gov/international/coc-docs/states.html')
state_soup = BeautifulSoup(state_res.content, 'lxml')
state_html = state_soup.find(text = 'Two-Letter State Abbreviations').find_next('table')
state = pd.DataFrame(pd.read_html(str(state_html))[0])
state.drop(state[state[0].isin(['AMERICAN SAMOA', 'GUAM', 'NORTHERN MARIANA IS', 'PUERTO RICO', 'VIRGIN ISLANDS'])].index, inplace = True)
state = state.reset_index(drop = True)[1]
state
```

```python
# remove the states with no polls, i.e. AK, DE, WY
state_without_polls = state
state_without_polls.drop(state_without_polls[state_without_polls.isin(['AK', 'DE', 'WY'])].index, inplace = True)
state_without_polls = state_without_polls.reset_index(drop = True)
state_without_polls
```

```python
# filter the dataframe to only include the state, month, year, democratic percentage, republican percentage, party winner, margin
polls_all = pd.DataFrame()
for i in range(len(dfs)):
    dfs[i]['state'] = [state_without_polls[i]] * len(dfs[i])
    month = []
    year = []
    for j in dfs[i]['Date administered']:
        if '–' in j:
            if j.split('–')[1][0] == ' ':
                month.append(j.split('–')[1][1:4])
            if j.split('–')[1][0] != ' ':
                month.append(j.split('–')[0][0:3])
        if '–' not in j:
            month.append(j.split(' ')[0][0:3])
        year.append(int(j.split(' ')[-1].split('[')[0]))
    dfs[i]['month'] = month
    dfs[i]['year'] = year
    dem_percentage = []
    for j in dfs[i]['%']:
        dem_percentage.append(float(str(j).split('%')[0]))
    dfs[i]['dem_percentage'] = dem_percentage
    gop_percentage = []
    for j in dfs[i]['%.1']:
        gop_percentage.append(float(str(j).split('%')[0]))
    dfs[i]['gop_percentage'] = gop_percentage
    winner = []
    margin = []
    for j in range(len(dfs[i])):
        if dfs[i]['dem_percentage'][j] > dfs[i]['gop_percentage'][j]:
            winner.append('D')
        if dfs[i]['dem_percentage'][j] < dfs[i]['gop_percentage'][j]:
            winner.append('R')
        if dfs[i]['dem_percentage'][j] == dfs[i]['gop_percentage'][j]:
            winner.append('E')
        margin.append(abs(dfs[i]['dem_percentage'][j] - dfs[i]['gop_percentage'][j]))
    dfs[i]['winner'] = winner
    dfs[i]['margin'] = margin
    polls_all = pd.concat([polls_all, dfs[i]], ignore_index = True)
```

```python
polls_dr = polls_all[['state', 'month', 'year', 'dem_percentage', 'gop_percentage', 'winner', 'margin']]
polls_dr
```

```python
# filter the dataframe to only include nov 2019 to nov 2020
polls_oneyear = polls_dr[(polls_dr['year'] == 2012) | ((polls_dr['year'] == 2011) & polls_dr['month'].isin(['Nov', 'Dec']))].reset_index(drop = True)
polls_oneyear
```

```python
# negative represents democratic, positive represents republican
party_margin = []
for i in range(len(polls_oneyear)):
    if polls_oneyear['winner'][i] == 'D':
        party_margin.append(-polls_oneyear['margin'][i])
    if any(polls_oneyear['winner'][i] in x for x in ['E', 'R']):
        party_margin.append(polls_oneyear['margin'][i])
polls_party_margin = polls_oneyear
polls_party_margin['2012_polls_margin'] = party_margin
polls_party_margin
```

```python
# find the unweighted mean
polls_mean = polls_party_margin.groupby('state', sort = False)[['2012_polls_margin']].mean().reset_index(drop = False)
polls_mean
```

```python
# export the dataframe to a csv file
polls_mean.to_csv('../../data/processed/2012_mean_polls.csv', index = False)
```
