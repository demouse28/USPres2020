```python
# import the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import collections
```

```python
# scrape the data on urban areas
res = requests.get('https://en.wikipedia.org/wiki/List_of_United_States_urban_areas')
soup = BeautifulSoup(res.content, 'lxml')
```

```python
soup
```

```python
urban = pd.read_html(str(soup.find('table', {'class': 'wikitable sortable'})))[0]
urban
```

```python
urban[['Area', 'States']] = urban['Name[Note 1]'].str.split(',', expand = True)
urban[['State_1', 'State_2', 'State_3', 'State_4']] = urban['States'].str.split('–', expand = True)
```

```python
urban = urban[['Area', 'State_1', 'State_2', 'State_3', 'State_4', 'Population(2020 Census)', 'Land Area(km2)', 'Density(Population / km2)']]
urban
```

```python
counts = collections.Counter(urban['State_1']) + collections.Counter(urban['State_2']) + collections.Counter(urban['State_3']) + collections.Counter(urban['State_4'])
counts
```

```python
dict(sorted(counts.items(), key = lambda item: item[1]))
```
